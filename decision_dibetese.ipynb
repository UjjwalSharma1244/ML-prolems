```
# extract attributes from dataframe
features=df.iloc[:,0:8].values
```

```
#now label
label=df.iloc[:,8].values
```

```
from sklearn.model_selection import train_test_split#septraining and testing data
```

```
X,x,Y,y=train_test_split(features,label,test_size=0.2)
#X is training data
# x is testing data
# Y is training answer /label
#y is testing answer /label
```

```
from sklearn.tree import DecisionTreeClassifier
```

```
clf=DecisionTreeClassifier()
```

```
trained=clf.fit(X,Y)
```

```
y_predict=trained.predict(x)
```

```
from sklearn.metrics import accuracy_score
```

```
accuracy_score(y,y_predict)
```

```
from sklearn.neighbors import KNeighborsClassifier
```

```
kclf=KNeighborsClassifier(n_neighbors=5)
```

```
kt=kclf.fit(X,Y)
```

```
out1=kt.predict(x)
```

```
accuracy_score(y,out1)
```
